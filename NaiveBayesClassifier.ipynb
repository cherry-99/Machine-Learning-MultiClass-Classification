```python
def mean_variance(x,y):
    no_of_attributes = x.shape[1]
    mean = np.zeros((no_of_attributes,3))
    variance = np.zeros((no_of_attributes,3))
    for i in range(no_of_attributes):
        count = [0,0,0]
        sums = [0,0,0]
        for j in range(len(y)):
            if y[j] == 0:
                count[0]+=1
                sums[0]=sums[0]+x[j][i]
            elif y[j] == 1:
                count[1]+=1
                sums[1]=sums[1]+x[j][i]
            elif y[j] == 2 :
                count[2]+=1
                sums[2]=sums[2]+x[j][i]
        mean[i]=[sums[0]/count[0] , sums[1]/count[1], sums[2]/count[2] ]
        sums = [0,0,0]
        for j in y:
            if j == 0:
                sums[0]+=pow((j-mean[i][0]),2)
            if j == 1:
                sums[1]+=pow((j-mean[i][1]),2)
            if j == 2:
                sums[2]+=pow((j-mean[i][2]),2)
        variance[i] = [sums[0]/(count[0]-1), sums[1]/(count[1]-1), sums[2]/(count[2]-1)]
    print("MEAN : shape = ",mean.shape)
    print(mean)
    print("VARIANCE : shape = ",variance.shape)
    print(variance)
    return mean,variance
```

```python
def prior_prob(y):
    d={}
    for i in y:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
    prior_probabilities = [(d[0]/len(y)),(d[1]/len(y)),(d[2]/len(y))]
    print("PRIOR_PROBABILITIES: ",prior_probabilities)
    return prior_probabilities
```

```python
def posterior_prob_per_class(mean,variance,test_val):
    pxc = [1,1,1]
#     for i in range(3):
#         for j in range(mean.shape[0]):
#             pxc[i] = pxc[i]*(1/sqrt(2*3.14*variance[j][i]))*exp(-0.5*(pow((test_val[j]-mean[j][i]),2))/(variance[j][i]))
    for i in range(0, 3):
        product = 1
        for j in range(0, mean.shape[0]):
            product = product * (1/sqrt(2*3.14*variance[j][i])) * exp(-0.5 * pow((test_val[j] - mean[j][i]),2)/variance[j][i])
        pxc[i] = product
    pxc = np.array(pxc)
    print("test data : ",test_val)
    print("Posterior Probabilities : ",pxc)
    return pxc
```

```python
def GNBpredict(X_train,Y_train,X_test):
    mean,variance = mean_variance(X_train,Y_train)
    prior_probabilities = prior_prob(Y_train)
    Y_pred=[]
    Y_pred_prob=[]
    for i in X_test:
        pcx = [1,1,1]
        prob = [1,1,1]
        posterior_probabilities = posterior_prob_per_class(mean,variance,i)
        total = 0
        for i in range(3):
            pcx[i] = prior_probabilities[i]*posterior_probabilities[i]
            total = total + pcx[i]
        print("Prior * Posterior : ",pcx)
        for i in range(3):
            prob[i] = pcx[i]/total
        prob = np.array(prob)
        print("Probabilities ",prob)
        prediction = int(prob.argmax())
        print("Prediction ",prediction)
        Y_pred.append(prediction)
    Y_pred = np.array(Y_pred)
    return Y_pred
```

```python
def generate_model_metrics(y_test,y_pred):
    x={"conf_matrix":metrics.confusion_matrix(y_test, y_pred),"Accuracy":metrics.accuracy_score(y_test,y_pred),"Precision":metrics.precision_score(y_test, y_pred, average="micro"),"Recall":metrics.recall_score(y_test, y_pred,average="micro"),"F1_Score":metrics.f1_score(y_test,y_pred,average="micro")}
    print("Confusion Matrix : \n",x["conf_matrix"])
    print("Accuracy of classifier : ",x["Accuracy"])
    print("Precision of classifier : ",x["Precision"])
    print("Sensitivity/Recall/TruePositiveRate of classifier : ",x["Recall"])
    print("F1 Score of classifier : ",x["F1_Score"])
```

```python
Y = samp1[target_column]
X = samp1.drop(columns=target_column)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30)
X_train = stats.zscore(X_train, axis=1, ddof=1)
X_test = stats.zscore(X_test, axis=1, ddof=1)
X_train = np.array(X_train)
Y_train = np.array(Y_train)
X_test = np.array(X_test)
Y_test = np.array(Y_test)
```

```python
Y_pred = GNBpredict(X_train, Y_train, X_train)
print("Confusion Matrix : \n",metrics.confusion_matrix(Y_train, Y_pred))
print("Classification Report : \n",metrics.classification_report(Y_train, Y_pred))
generate_model_metrics(Y_train, Y_pred)
print("Y_test : ",Y_train)
print("Y_pred : ",Y_pred)
```
